```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, color='blue', marker='o', linestyle='-')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()
```

```python
x = ["red", "blue", "green", "orange", "purple"]
y = [5, 7, 3, 8, 4]
plt.bar(x, y, color=['red', 'blue', 'green', 'orange', 'purple'])
plt.xlabel("Colors")    
plt.ylabel("Values")
plt.title("Simple Bar Chart")
plt.show()
```

```python
x = ["red", "blue", "green", "orange", "purple"]
y = [5, 7, 3, 8, 4]

plt.pie(y, labels=x, autopct='%1.1f%%', startangle=140)
plt.title("Simple Pie Chart")
plt.show()
```

```python
import numpy as np

x = np.random.rand(1000, 1)

plt.hist(x, bins=30, color='blue', edgecolor='black', alpha=0.7)
plt.show()
```

```python
x = np.random.rand(65, 65)
plt.imshow(x, cmap='viridis')
plt.colorbar()
plt.title("Heatmap Example")
plt.show()
```
