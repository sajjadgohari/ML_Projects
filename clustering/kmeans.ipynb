```python
from sklearn.datasets import make_blobs

x, y = make_blobs(n_samples=300, n_features=2, centers=3)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x[:, 0], x[:, 1], color='b')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
```

```python
center = kmeans.cluster_centers_
lambdabels = kmeans.labels_
```

```python
plt.scatter(x[:, 0], x[:, 1], c=lambdabels, cmap='rainbow')
plt.scatter(center[:, 0], center[:, 1], marker='*', color='black', s=200)
plt.show()
```

```python
# we see that the kmeans algorithm has successfully identified the clusters and their centers
```
